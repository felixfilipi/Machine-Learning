```python
#import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
ds= pd.read_csv('datasets/housingprice.txt', delimiter =',',header=None) 
ds.columns = ["Size", "Bedrooms", "Price"]
```

```python
#print data
ds
```

```python
#split train and target
train=ds[['Size','Bedrooms']]
target=ds['Price']
```

```python
#build model
from sklearn.linear_model import LinearRegression 
regressor= LinearRegression() 
regressor.fit(train,target) 
```

```python
#Asumsikan anda menjual rumah dan ingin mengetahui harga pasaran rumah
Size=input("Size= ")
Bedrooms=input("Bedrooms= ")
```

```python
#Give predict result
int(regressor.predict([[Size,Bedrooms]]))
```
