```python
#Decision Tree Learning
import pandas as pd

ds = pd.read_csv('datasets/lie-dataset.csv')
```

```python
ds
```

```python
x = ds[['HeartRate','BlinkFreq','EyeContact']]
y = ds['Class']
```

```python
from sklearn.tree import DecisionTreeClassifier

dtl=DecisionTreeClassifier()
test=dtl.fit(x,y)
```

```python
ht=input('Heart Rate = ')
bf=input('Blink Freq = ')
ec=input('Eye Contact = ')
```

```python
test.predict([[ht,bf,ec]])
```

```python
#Viualize dtl process
from sklearn.tree import export_graphviz
export_graphviz(
    dtl,
    out_file = "models/visualize.dot",
    feature_names = ['HeartRate','BlinkFreq','EyeContact'],
    class_names = ['Lying', 'Not-Lying'],
    rounded= True,
    filled =True
)
```
