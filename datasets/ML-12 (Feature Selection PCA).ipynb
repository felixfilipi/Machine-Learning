<h3>Import Dataset</h3>

```python
import pandas as pd

ds=pd.read_csv('datasets/listings.csv')

ds
```

<h3>Apply pandas factorize ke data</h3>

```python
ds=ds.apply(lambda ds : pd.factorize(ds)[0])
ds
```

<h2>1. Discuss to find the correlation and covariance of the data (the variables used can be categorical vs categorical, categorical vs numeric, or numeric vs numeric variables) </h2>

```python
ds.corr(method='pearson')
```

```python
ds.cov()
```

 <h2>2. Normalized the data using Standardization or Range Scaling method. (MaxAbsScaler)</h2>

```python
from sklearn.preprocessing import MaxAbsScaler

scaled=MaxAbsScaler().fit_transform(ds)
data=pd.DataFrame(scaled,columns=ds.columns)
data
```

<h2>3. Do feature selection and extraction. </h2>

<h3>Untuk mengetahui berapa dimensi yang kita perlukan kita dapat melihat variance tiap attribute </h3>

```python
from sklearn.decomposition import PCA

pca=PCA()
data=pca.fit_transform(data)
pca.explained_variance_ratio_
```

<h3>Setelah mengecek variancenya, kita dapat tentukan dimension reduction yang tepat (saya ambil 12 attribute saja dan membuang 4 attribute terakhir)</h3>

```python
pca=PCA(n_components=12)
data=pca.fit_transform(data)
pd.DataFrame(data,columns=ds.columns[0:12])
```
